```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
df = pd.read_csv('cardio_train.csv', sep=';')

# Realizar anlisis exploratorio de datos
print(df.head())
print(df.describe())

# Visualizar la distribucin de las clases
sns.countplot(x='cardio', data=df)
plt.title('Distribucin de Clases')
plt.show()

# Visualizar la distribucin de las variables numricas
df.hist(bins=50, figsize=(20, 15))
plt.show()

# Visualizar la correlacin entre las variables
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
